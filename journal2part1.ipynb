```python
import matplotlib.pyplot as plt
import numpy as np

inc = 1/100
x = np.linspace(0, 1, 100)
```

```python
def exp(x):
    return np.exp(x)     # natural log ** x
```

```python
y = exp(x)
plt.plot(x, y)
plt.xlabel("Time [milliseconds]")
plt.ylabel("Awesomeness (heck yeah)")
```

```python
c = np.linspace(0, 2, 1000)
d = exp(c)
plt.plot(c, d)
plt.xlabel("X")
plt.ylabel("X**X")
```
